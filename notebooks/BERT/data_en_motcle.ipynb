```python
import spacy
from collections import Counter


l=[df_non_coherent_questions_1, df_coherent_questions_1, df_non_coherent_questions_2, df_coherent_questions_2, df_non_coherent_questions_3, df_coherent_questions_3, df_non_coherent_questions_4, df_coherent_questions_4, df_non_coherent_questions_5, df_coherent_questions_5]

l_name=['df_non_coherent_questions_1', 'df_coherent_questions_1', 'df_non_coherent_questions_2', 'df_coherent_questions_2', 'df_non_coherent_questions_3', 'df_coherent_questions_3', 'df_non_coherent_questions_4', 'df_coherent_questions_4', 'df_non_coherent_questions_5', 'df_coherent_questions_5']

nlp = spacy.load("en_core_web_sm")

def extract_keywords(text):
    doc = nlp(text)
    keywords = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return keywords
```

```python
for i,n in zip(l, l_name):
    df_key = i.copy()
    
    df_key['keywords'] = df_key['answer'].apply(extract_keywords)

    all_keywords_list = [keyword for sublist in df_key['keywords'] for keyword in sublist]


    keyword_counts = Counter(all_keywords_list)

    print(f"\nTop 10 des mots-clés les plus fréquents {n} :")
    print(keyword_counts.most_common(10))
```
